```python
def row_wise_feat_engi(df):
    df = df.copy()
    new_features = {}

    new_features['row_mean'] = df[feature_list].mean(axis=1)
    new_features['row_std'] = df[feature_list].std(axis=1)
    new_features['row_max'] = df[feature_list].max(axis=1)
    new_features['row_min'] = df[feature_list].min(axis=1)
    new_features['row_sum'] = df[feature_list].sum(axis=1)

    for i in tqdm(range(19)):
        nth = round(0.05 + i * 0.05, 2)
        new_features['row_{}_quantile'.format(nth)] = df[feature_list].quantile(q=nth, axis=1)

    new_feats_df = pd.DataFrame(new_features, index=df.index)
    result_df = pd.concat([df, new_feats_df], axis=1)

    return result_df.copy()
```

### Nonlinear transformations

```python
top_20_features_list = ['X272',
 'X758',
 'X757',
 'X772',
 'X17',
 'X607',
 'X756',
 'X443',
 'X648',
 'X567',
 'X674',
 'X86',
 'X25',
 'X767',
 'X685',
 'X264',
 'X752',
 'X413',
 'X464',
 'X614']
```

```python
def nonlinear_feat_engi(df):
    df = df.copy()
    new_features = {}

    for f in tqdm(top_20_features_list):
        new_features["{}_percentile_rank".format(f)] = df[f].rank(pct=True)
        new_features["{}_square".format(f)] = df[f].apply(lambda x: x**2)
        new_features["{}_cube".format(f)] = df[f].apply(lambda x: x**3)
        new_features["{}_sqrt".format(f)] = df[f].apply(lambda x: np.sqrt(np.abs(x)))
        new_features["{}_log1p".format(f)] = df[f].apply(lambda x: np.log1p(np.abs(x)))
        new_features["{}_exp".format(f)] = df[f].apply(lambda x: np.exp(x))
        new_features["{}_sin".format(f)] = df[f].apply(lambda x: np.sin(x))
        new_features["{}_cos".format(f)] = df[f].apply(lambda x: np.cos(x))
        new_features["{}_tanh".format(f)] = df[f].apply(lambda x: np.tanh(x))
        new_features["{}_abs".format(f)] = df[f].apply(lambda x: np.abs(x))
    
    new_feats_df = pd.DataFrame(new_features, index=df.index)
    result_df = pd.concat([df, new_feats_df], axis=1)

    return result_df.copy()
```

```python
train_df = nonlinear_feat_engi(train_df)
print(train_df.shape)
train_df.head()
```

```python
test_df = nonlinear_feat_engi(test_df)
print(test_df.shape)
test_df.head()
```

### Interaction feature engineering

```python
def interaction_feat_engi(df):
    df = df.copy()
    new_features = {}

    for f1, f2 in combinations(top_20_features_list, 2):
        new_features[f'{f1}_{f2}_prod'] = df[f1] * df[f2]
        new_features[f'{f1}_{f2}_sum'] = df[f1] + df[f2]
        new_features[f'{f1}_{f2}_diff'] = df[f1] - df[f2]
        new_features[f'{f1}_{f2}_ratio'] = df[f1] / (df[f2] + 1e-5)
        new_features[f'{f1}_{f2}_max'] = df[[f1, f2]].max(axis=1)
        new_features[f'{f1}_{f2}_min'] = df[[f1, f2]].min(axis=1)
        new_features[f'{f1}_{f2}_absdiff'] = np.abs(df[f1] - df[f2])

    new_feats_df = pd.DataFrame(new_features, index=df.index)
    result_df = pd.concat([df, new_feats_df], axis=1)

    return result_df.copy()
```

```python
train_df = interaction_feat_engi(train_df)
print(train_df.shape)
train_df.head()
```

# Ensemble

```python
import datetime

# Get the current time in the local timezone
current_time_local = datetime.datetime.now()
print(f"Current time (local): {current_time_local}"[:-7])
```

```python
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
```

```python
def ensemble():
    X = train_df.drop(columns=["label"])
    y = train_df["label"]
    X_test = test_df.drop(columns=["label"])
    
    n_splits = 5
    tscv = TimeSeriesSplit(n_splits=n_splits)

    lgb_params = {
        "num_iterations": 5000,
        "early_stopping_round": 100,
        "device": "gpu",
        "verbosity": -1,
        'learning_rate': 0.01,
        'max_depth': 6,
        'min_data_in_leaf': 10,
        'num_leaves': 130,
        'min_gain_to_split': 0.4,
        'lambda_l1': 5.119331306798383,
        'lambda_l2': 0.0,
        'feature_fraction': 0.7
    }

    xgb_params = {
        'tree_method': 'gpu_hist',
        'n_estimators': 5000,
        'verbosity': 0,
        'early_stopping_rounds': 100,
        'learning_rate': 0.06199170763007776,
        'max_depth': 10,
        'subsample': 0.6,
        'colsample_bytree': 0.6,
        'reg_alpha': 0.0,
        'reg_lambda': 0.0,
        'min_child_weight': 200.0,
        'min_split_loss': 199.99999999999997
    }

    catboost_params = {
        'iterations': 5000,
        'verbose': 0,
        'task_type': 'GPU',
        'use_best_model': True,
        'early_stopping_rounds': 100,
        'learning_rate': 0.05977842545724329,
        'depth': 4,
        'l2_leaf_reg': 43,
        'bagging_temperature': 1.3883525584252365,
        'random_strength': 32.23480222920519
    }

    lgb_oof_preds = np.full(len(train_df), np.nan)
    xgb_oof_preds = np.full(len(train_df), np.nan)
    catboost_oof_preds = np.full(len(train_df), np.nan)

    lgb_preds = np.full((len(test_df), n_splits), np.nan)
    xgb_preds = np.full((len(test_df), n_splits), np.nan)
    catboost_preds = np.full((len(test_df), n_splits), np.nan)

    for i, (train_idx, val_idx) in enumerate(tscv.split(train_df)):
        print(f"{i} fold --------")
        X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
        X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]

        lgb_model = LGBMRegressor(**lgb_params)
        xgb_model = XGBRegressor(**xgb_params)
        catboost_model = CatBoostRegressor(**catboost_params)

        lgb_model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_val, y_val)])
        xgb_model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_val, y_val)], verbose=0)
        catboost_model.fit(X_train, y_train, eval_set=(X_val, y_val))

        lgb_oof_preds[val_idx] = lgb_model.predict(X_val)
        xgb_oof_preds[val_idx] = xgb_model.predict(X_val)
        catboost_oof_preds[val_idx] = catboost_model.predict(X_val)

        lgb_preds[:, i] = lgb_model.predict(X_test)
        xgb_preds[:, i] = xgb_model.predict(X_test)
        catboost_preds[:, i] = catboost_model.predict(X_test)
    
    lgb_mask = ~np.isnan(lgb_oof_preds)
    lgb_score = np.corrcoef(y[lgb_mask], lgb_oof_preds[lgb_mask])[0, 1]
    
    xgb_mask = ~np.isnan(xgb_oof_preds)
    xgb_score = np.corrcoef(y[xgb_mask], xgb_oof_preds[xgb_mask])[0, 1]

    catboost_mask = ~np.isnan(catboost_oof_preds)
    catboost_score = np.corrcoef(y[catboost_mask], catboost_oof_preds[catboost_mask])[0, 1]

    print(f"LGB score: {lgb_score}")
    print(f"XGB score: {xgb_score}")
    print(f"Catboost score: {catboost_score}")

    score_sum = lgb_score + xgb_score + catboost_score

    ensemble_oof_preds = (lgb_oof_preds * (lgb_score / score_sum)) + (xgb_oof_preds * (xgb_score / score_sum)) + (catboost_oof_preds * (catboost_score / score_sum))
    ensemble_mask = ~np.isnan(ensemble_oof_preds)
    ensemble_score = np.corrcoef(y[ensemble_mask], ensemble_oof_preds[ensemble_mask])[0, 1]

    print(f"Ensemble score: {ensemble_score}")

    lgb_preds_mean = lgb_preds.mean(axis=1)
    xgb_preds_mean = xgb_preds.mean(axis=1)
    catboost_preds_mean = catboost_preds.mean(axis=1)

    ensemble_preds = (lgb_preds_mean * (lgb_score / score_sum)) + (xgb_preds_mean * (xgb_score / score_sum)) + (catboost_preds_mean * (catboost_score / score_sum))

    return ensemble_preds
```

```python
ensemble_preds = ensemble()
```
